# 1.3 Deskriptivna statistika z grafi

V tem razdelku pripraviva opisno statistiko in grafične prikaze **za vsako spremenljivko**:
- numerične spremenljivke: *mean ± SD* ali *mediana (Q1–Q3)*, **min–max**, ter histogram (porazdelitev),
- kategorialne spremenljivke: **n (%)** po kategorijah ter stolpčni diagram (deleži),
- pri vsaki spremenljivki - kratka interpretacija (porazdelitev, odstopanja, posebnosti).


## 1) Uvoz podatkov in priprava pomožnih funkcij

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

CSV_PATH = "CVD_cleaned.csv"  # prilagodi pot po potrebi
df = pd.read_csv(CSV_PATH)

df.shape, df.columns.tolist()
```

```python
# Ločimo numerične in kategorialne spremenljivke
num_cols = df.select_dtypes(include="number").columns.tolist()
cat_cols = [c for c in df.columns if c not in num_cols]

num_cols, cat_cols
```

```python
def numeric_summary(s: pd.Series) -> dict:
    """Vrne osnovne statistike za numerično spremenljivko."""
    s = s.dropna()
    q1 = float(s.quantile(0.25))
    med = float(s.quantile(0.50))
    q3 = float(s.quantile(0.75))
    iqr = q3 - q1
    out_hi = float(q3 + 1.5 * iqr)
    out_lo = float(q1 - 1.5 * iqr)
    
    return {
        "n": int(s.shape[0]),
        "mean": float(s.mean()),
        "sd": float(s.std(ddof=1)),
        "median": med,
        "q1": q1,
        "q3": q3,
        "min": float(s.min()),
        "max": float(s.max()),
        "skew": float(s.skew()),
        "outlier_rule_lo": out_lo,
        "outlier_rule_hi": out_hi
    }

def choose_report_style(skew: float, threshold: float = 0.5) -> str:
    """Enostavno pravilo: če je |skew| < threshold -> mean±SD, sicer median(IQR)."""
    return "mean±SD" if abs(skew) < threshold else "median (Q1–Q3)"

def categorical_table(s: pd.Series) -> pd.DataFrame:
    """Tabela n (%) za kategorialno spremenljivko."""
    counts = s.value_counts(dropna=False)
    perc = counts / counts.sum() * 100
    out = pd.DataFrame({"n": counts, "%": perc.round(2)})
    out.index.name = "kategorija"
    return out

def interpret_numeric(name: str, st: dict) -> str:
    """Kratek komentar za numerično spremenljivko."""
    skew = st["skew"]
    if abs(skew) < 0.5:
        shape = "približno simetrična"
    elif skew > 0:
        shape = "desno asimetrična (daljši rep proti višjim vrednostim)"
    else:
        shape = "levo asimetrična (daljši rep proti nižjim vrednostim)"
        
    # približen indikator izstopajočih vrednosti po 1.5*IQR pravilu
    outliers_hi = int((df[name] > st["outlier_rule_hi"]).sum())
    outliers_lo = int((df[name] < st["outlier_rule_lo"]).sum())
    out_n = outliers_hi + outliers_lo
    out_share = out_n / st["n"] * 100
    
    style = choose_report_style(skew)
    if style == "mean±SD":
        center = f"povprečje {st['mean']:.2f} ± {st['sd']:.2f}"
    else:
        center = f"mediana {st['median']:.2f} (Q1–Q3: {st['q1']:.2f}–{st['q3']:.2f})"
    
    return (
        f"- Porazdelitev je {shape}.\n"
        f"- Osrednja vrednost ({style}): {center}.\n"
        f"- Razpon: {st['min']:.2f}–{st['max']:.2f}.\n"
        f"- Po pravilu 1.5×IQR je približno {out_n} potencialnih izstopajočih vrednosti ({out_share:.2f}%)."
    )

def interpret_categorical(name: str, tab: pd.DataFrame) -> str:
    """Kratek komentar za kategorialno spremenljivko."""
    k = tab.shape[0]
    top = tab.sort_values("n", ascending=False).iloc[0]
    top_cat = tab.sort_values("n", ascending=False).index[0]
    top_share = float(top["%"])
    
    imbalance = ""
    if top_share >= 70:
        imbalance = " Porazdelitev je močno neuravnotežena."
    elif top_share >= 50:
        imbalance = " Porazdelitev je precej neuravnotežena."
        
    return (
        f"- Spremenljivka ima {k} kategorij.\n"
        f"- Najpogostejša kategorija je **{top_cat}** ({int(top['n'])} zapisov; {top_share:.2f}%)."
        f"{imbalance}"
    )
```

## 2) Tabela opisne statistike za vse spremenljivke

Spodaj ustvarimo *skupno tabelo* opisne statistike za:
- numerične spremenljivke (izbran stil poročanja + min–max),
- kategorialne spremenljivke (kategorije z n (%)).


```python
rows = []

# Numerične spremenljivke
for c in num_cols:
    st = numeric_summary(df[c])
    style = choose_report_style(st["skew"])
    if style == "mean±SD":
        summary = f"{st['mean']:.2f} ± {st['sd']:.2f}"
    else:
        summary = f"{st['median']:.2f} ({st['q1']:.2f}–{st['q3']:.2f})"
        
    rows.append({
        "Spremenljivka": c,
        "Tip": "numerična",
        "Kategorija": "",
        "n": st["n"],
        "%": "",
        "Osnovna statistika": summary,
        "Min": st["min"],
        "Max": st["max"],
        "Skew": st["skew"],
        "Priporočen zapis": style
    })

# Kategorialne spremenljivke
for c in cat_cols:
    tab = categorical_table(df[c])
    for lvl, r in tab.iterrows():
        rows.append({
            "Spremenljivka": c,
            "Tip": "kategorialna",
            "Kategorija": lvl,
            "n": int(r["n"]),
            "%": float(r["%"]),
            "Osnovna statistika": "n (%)",
            "Min": "",
            "Max": "",
            "Skew": "",
            "Priporočen zapis": ""
        })

desc_table = pd.DataFrame(rows)

# Prikaz (najprej numerične, nato kategorialne)
desc_table_numeric = desc_table[desc_table["Tip"]=="numerična"].copy()
desc_table_cat = desc_table[desc_table["Tip"]=="kategorialna"].copy()

desc_table_numeric, desc_table_cat.head(10)
```

```python
# Opcijsko: shranimo tabelo v CSV (uporabno za poročilo)
OUT_TABLE = "1_3_opisna_statistika.csv"
desc_table.to_csv(OUT_TABLE, index=False)
OUT_TABLE
```

## 3) Numerične spremenljivke – statistike, grafi, interpretacija

Za vsako numerično spremenljivko:
- izpišemo osnovne statistike,
- narišemo **histogram**,
- kratka interpretacija (porazdelitev, izstopajoče vrednosti).


```python
for c in num_cols:
    st = numeric_summary(df[c])
    style = choose_report_style(st["skew"])
    
    print(f"\n {c}")
    print(f"n = {st['n']}")
    print(f"Priporočen zapis: {style}")
    print(f"mean ± SD: {st['mean']:.2f} ± {st['sd']:.2f}")
    print(f"median (Q1–Q3): {st['median']:.2f} ({st['q1']:.2f}–{st['q3']:.2f})")
    print(f"min–max: {st['min']:.2f}–{st['max']:.2f}")
    print(f"skew: {st['skew']:.2f}")
    
    # Graf: histogram
    plt.figure()
    plt.hist(df[c].dropna(), bins=40)
    plt.title(f"Histogram: {c}")
    plt.xlabel(c)
    plt.ylabel("Frekvenca")
    plt.show()
    
    # Interpretacija
    print("Interpretacija:")
    print(interpret_numeric(c, st))
```

## 4) Kategorialne spremenljivke – n (%), grafi, interpretacija

Za vsako kategorialno spremenljivko:
- izpišemo tabelo **n (%)**,
- narišemo **stolpčni diagram deležev**,
- kratka interpretacija (najpogostejša kategorija, neuravnoteženost).


```python
for c in cat_cols:
    tab = categorical_table(df[c])
    
    print(f"\n {c}")
    display(tab)
    
    # Graf: deleži (%), sortirano padajoče
    tab_sorted = tab.sort_values("%", ascending=False)
    plt.figure()
    plt.bar(tab_sorted.index.astype(str), tab_sorted["%"].values)
    plt.title(f"Deleži (%): {c}")
    plt.xlabel("Kategorija")
    plt.ylabel("Delež (%)")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
    
    # Interpretacija
    print("Interpretacija:")
    print(interpret_categorical(c, tab))
```

---

- Numerične spremenljivke sva poročala kot **mean ± SD** pri približno simetričnih porazdelitvah (|skew| < 0.5), sicer kot **mediana (Q1–Q3)**.  
- Kategorialne spremenljivke sva poročala kot **n (%)**.  

```python
# === 1.3: Export vseh grafov v PNG + priprava markdown snippetov za README ===
from pathlib import Path
import re
import matplotlib.pyplot as plt

# Kam shranjujemo slike (v repo priporočam npr. "figures/1_3")
OUT_DIR = Path("figures/1_3")
OUT_DIR.mkdir(parents=True, exist_ok=True)

def slugify(name: str) -> str:
    """Safe ime datoteke: črke/številke/_; ostalo -> _"""
    name = re.sub(r"[^0-9a-zA-Z]+", "_", name)
    return name.strip("_")

md_lines = []
md_lines.append("## 1.3 Grafi (avtomatsko generirano)\n")

# --- Numerične spremenljivke: histogram + boxplot ---
for c in num_cols:
    fname_base = slugify(c)

    # Histogram
    plt.figure()
    plt.hist(df[c].dropna(), bins=40)
    plt.title(f"Histogram: {c}")
    plt.xlabel(c)
    plt.ylabel("Frekvenca")
    hist_path = OUT_DIR / f"hist_{fname_base}.png"
    plt.tight_layout()
    plt.savefig(hist_path, dpi=200)
    plt.close()

    # Boxplot (dobro za outlierje)
    plt.figure()
    plt.boxplot(df[c].dropna(), vert=True)
    plt.title(f"Boxplot: {c}")
    plt.ylabel(c)
    box_path = OUT_DIR / f"box_{fname_base}.png"
    plt.tight_layout()
    plt.savefig(box_path, dpi=200)
    plt.close()

    md_lines.append(f"### {c}\n")
    md_lines.append(f"![Histogram {c}]({hist_path.as_posix()})\n")
    md_lines.append(f"![Boxplot {c}]({box_path.as_posix()})\n")

# --- Kategorialne spremenljivke: barplot deležev ---
for c in cat_cols:
    tab = categorical_table(df[c]).sort_values("%", ascending=False)
    fname_base = slugify(c)

    plt.figure()
    plt.bar(tab.index.astype(str), tab["%"].values)
    plt.title(f"Deleži (%): {c}")
    plt.xlabel("Kategorija")
    plt.ylabel("Delež (%)")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()

    bar_path = OUT_DIR / f"bar_{fname_base}.png"
    plt.savefig(bar_path, dpi=200)
    plt.close()

    md_lines.append(f"### {c}\n")
    md_lines.append(f"![Deleži {c}]({bar_path.as_posix()})\n")

# --- Shrani markdown snippet za README ---
snippets_path = OUT_DIR / "README_snippets.md"
snippets_path.write_text("\n".join(md_lines), encoding="utf-8")

print(f"✅ Shranjeno v: {OUT_DIR.resolve()}")
print(f"✅ README snippet: {snippets_path.resolve()}")
```
